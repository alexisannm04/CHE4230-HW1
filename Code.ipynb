```python
print("Problem 3")

def Prime_Check(num):
    if num<=1:
        return False
    
    for i in range(2,int(num**0.5)+1):
        if num % i == 0:
            return False
        
    else:
        return True
    
print("b)")
primes_below_100 = []
for x in range (2,100):
    if Prime_Check(x):
        primes_below_100.append(x)
print(primes_below_100)

def primes_below(n):
    primes=[]
    for num in range(2,n):
        if Prime_Check(num):
            primes.append(num)
    return primes

print("d)")
primes_below_1000=primes_below(1000)
print(primes_below_1000)
```

```python
print("Problem 4")

import numpy as np

def Create_3D_Array(n,data):

    if data == "zeroes":
        return np.zeros([n,3,3])

    elif data == "random":
        return np.random.randint(101,size=(n,3,3))
    
print("b)")

zero_array=Create_3D_Array(10,"zeroes")
print(zero_array)

print("c)")

random_array=Create_3D_Array(5,"random")
print(random_array)
```

```python
print("Problem 5")

import numpy as np
A = np.matrix([
    [3,4,-5],
    [0,5,1],
    [1,3,6]
])

B=np.matrix([
    [10],
    [8],
    [-5]
])

X = np.dot(np.linalg.inv(A),B)
print(X)

```

```python
print("Problem 6")
import matplotlib.pyplot as plt
import numpy as np

y= 0.1*x**3-x**2+5
x=np.linspace(-5,5,5)
fig, ax= plt.subplots(figsize=(5,2.7),layout='constrained')
ax.plot(x,(y))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Function: y=0.1x^3-x^2+5")

#Max and Minimum
y_max=np.max(y)
x_max=x[np.argmax(y)]
y_min=np.min(y)
x_min=x[np.argmin(y)]

plt.scatter(x_max, y_max)
plt.scatter(x_min, y_min)

plt.text(x_max, y_max, f" Max ({x_max:.2f}, {y_max:.2f})")
plt.text(x_min, y_min, f" Min ({x_min:.2f}, {y_min:.2f})")




```
